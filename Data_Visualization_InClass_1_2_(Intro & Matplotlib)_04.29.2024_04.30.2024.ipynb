```python
x = np.arange(-50, 50)
y1 = np.power(x, 2)
y2 = np.power(x, 3)
y3 = np.sin(x)
y4 = np.cos(x)
y5 = np.tan(x)
y6 = np.tanh(x)
y7 = np.sinh(x)
y8 = np.cosh(x)
y9 = np.exp(x)

fig_1, ax_1 = plt.subplots(nrows= 3, ncols= 3, figsize = (20,20))

ax_1[0, 0].plot(x, y1, "tab:blue")
ax_1[0, 0].set_title("Square Function")
ax_1[0, 0].set_xlabel("X", fontsize = 18)
ax_1[0, 0].set_ylabel("Y", fontsize = 18)

ax_1[0, 1].plot(x, y2, "tab:blue")
ax_1[0, 1].set_title("Cubic Function")
ax_1[0, 1].set_xlabel("X", fontsize = 18)
ax_1[0, 1].set_ylabel("Y", fontsize = 18)

ax_1[0, 2].plot(x, y3, "tab:blue")
ax_1[0, 2].set_title("Sine Function")
ax_1[0, 2].set_xlabel("X", fontsize = 18)
ax_1[0, 2].set_ylabel("Y", fontsize = 18)

ax_1[1, 0].plot(x, y4, "tab:blue")
ax_1[1, 0].set_title("Cosine Function")
ax_1[1, 0].set_xlabel("X", fontsize = 18)
ax_1[1, 0].set_ylabel("Y", fontsize = 18)

ax_1[1, 1].plot(x, y5, "tab:blue")
ax_1[1, 1].set_title("Tangent Function")
ax_1[1, 1].set_xlabel("X", fontsize = 18)
ax_1[1, 1].set_ylabel("Y", fontsize = 18)

ax_1[1, 2].plot(x, y6, "tab:blue")
ax_1[1, 2].set_title("Hyperbolic Tangent Function")
ax_1[1, 2].set_xlabel("X", fontsize = 18)
ax_1[1, 2].set_ylabel("Y", fontsize = 18)

ax_1[2, 0].plot(x, y7 ,"tab:blue")
ax_1[2, 0].set_title("Hyperbolic Sine Function")
ax_1[2, 0].set_xlabel("X", fontsize = 18)
ax_1[2, 0].set_ylabel("Y", fontsize = 18)


ax_1[2, 1].plot(x, y8, "tab:blue")
ax_1[2, 1].set_title("Hyperbolic Cosine Function")
ax_1[2, 1].set_xlabel("X", fontsize = 18)
ax_1[2, 1].set_ylabel("Y", fontsize = 18)


ax_1[2, 2].plot(x, y9, "tab:blue")
ax_1[2, 2].set_title("Exp Function")
ax_1[2, 2].set_xlabel("X", fontsize = 18)
ax_1[2, 2].set_ylabel("Y", fontsize = 18)

plt.tight_layout()
plt.show()
```

```python
fig, ax = plt.subplots()

ax.bar(tips.sex, tips.tip)
plt.show()
```

```python
avg_tips = tips.groupby("sex")["tip"].mean()

fig, ax = plt.subplots()
ax.bar(avg_tips.index, avg_tips.values)
ax.set_title("Average Tip by Sex")
ax.set_ylabel("Average Tip")
ax.set_xlabel("Sex")
plt.show()
```

```python
avg_tips = tips.groupby("sex")["tip"].sum()

fig, ax = plt.subplots()
ax.bar(avg_tips.index, avg_tips.values)
ax.set_title("Total Tip by Sex")
ax.set_ylabel("Total Tip")
ax.set_xlabel("Sex")
plt.show()
```

```python
avg_tips = tips.groupby("sex")["tip"].sum()

fig, ax = plt.subplots()
ax.pie(avg_tips.values, labels=avg_tips.index)
ax.set_title("Total Tip by Sex")
ax.set_ylabel("Total Tip")
ax.set_xlabel("Sex")
plt.show()
```

```python
x = np.array([25,35,25,15])
mylabels = ["Apples", "Bananas", "Cherries", "Melon"]
color_list = ["gold", "yellowgreen","lightcoral", "lightskyblue"]
myexplode = [0,0,0,0.5]

plt.figure(figsize=(10, 8))

plt.pie(x,
        labels = mylabels,
        explode= myexplode, 
        labeldistance = 0.7, 
        autopct = "%.2f",
        startangle=180,
        shadow= True, 
        colors= color_list,
        )
plt.legend(loc =(1))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sex_counts = tips.sex.value_counts()
color_list= ["lightskyblue", "lightcoral"]
myexplode = [0.1, 0]

plt.pie(sex_counts,
        labels = sex_counts.index,
        explode= myexplode, 
        labeldistance = 0.7, 
        autopct = "%1.1f",
        startangle=140,
        shadow= True, 
        colors= color_list,
        )
plt.title("Tips Dataset - Meal Counts by Sex")
plt.legend(loc =(1))
plt.show()
```
