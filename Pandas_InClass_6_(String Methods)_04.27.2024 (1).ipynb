# <font color='green'> <b>Text Data - String Methods</b><font color='black'>
    
**str.lower():** Converts a string to lowercase

**str.upper():** Converts a string to uppercase

**str.capitalize():** Converts the first character to uppercase

**str.title():** Converts the first character of each word to uppercase

**str.swapcase():** Converts uppercase letters to lowercase and lowercase letters to uppercase

**str.isalpha():** Returns True or False by checking whether all characters are in the alphabet.

**str.isnumeric():** Checks whether all characters contain numeric values and returns True or False

**str.isalnum():** Returns True or False by checking whether all characters contain letters or numeric values

**str.endswith():** Returns True or False by checking that a string ends with the specified value

**str.startswith():** Returns True or False by checking that a string starts with the specified value

**str.contains():** Checks whether a substring is contained in a string, returning True or False for each element

**str.strip():** Removes leading and trailing spaces from a string and returns a trimmed version

**str.replace():** Returns a string with a specified value replaced by another value.

**str.split():** Splits a string at the specified delimiter and returns a list

**str.find():** Searches for a specified value in a string and returns the index of its location

**str.findall():** Returns a list of all instances in a pattern

**str.join():** Converts the elements of an iterable to a string
