```python
age = np.arange(25, 44, 2.5)
salary = np.arange(40000, 76000, 5000)
salary_2 = [45372, 46852, 69401, 56790, 73000, 75000, 80000, 83545]
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"b:", label= "SALARY")
plt.plot(age, salary_2, "r-", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend();
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"b:")
plt.plot(age, salary_2, "r-")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend();
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"bo", label= "SALARY")
plt.plot(age, salary_2, "r*", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend();
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"bo", label= "SALARY")
plt.plot(age, salary_2, "r*", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(loc = "lower right")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"bo", label= "SALARY")
plt.plot(age, salary_2, "r*", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(loc = "right")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"bo", label= "SALARY")
plt.plot(age, salary_2, "r*", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor = (1,0.2))
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"bo", label= "SALARY")
plt.plot(age, salary_2, "r*", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.5,0.5))
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary, label= "SALARY")
plt.plot(age, salary_2, label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.5,0.5))
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.plot(age, salary,"r", label= "SALARY")
plt.plot(age, salary_2, "y", label= "SALARY_2")
plt.xlabel("AGE")
plt.ylabel("SALARY")
plt.title("SALARY BY AGE")
plt.legend(bbox_to_anchor = (0.5,0.5))
plt.show()
```
