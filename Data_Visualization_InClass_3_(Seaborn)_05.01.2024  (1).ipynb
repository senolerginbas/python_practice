## Categorical Plots

### Countplot

```python
sns.countplot(data = tips, x = "day");
```

```python
sns.countplot(data = mpg, x = "cylinders");
```

```python
fig, ax = plt.subplots()
ax = sns.countplot(data = tips, x = "day");

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x() + 0.3, p.get_height() + 0.6))
```

```python
print(ax.patches)
```

```python
for p in ax.patches:
    print(p)
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

ax_tip = sns.countplot(data = tips, x = "day", hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+1))
```

```python
fig, ax = plt.subplots(figsize = (10, 5))

ax_tip = sns.countplot(data = tips, x = "day", hue = "sex")

for p in ax.patches:
    ax.annotate((p.get_height()), (p.get_x()+0.15, p.get_height()+1))
    ax.text(p.get_x()+0.07, p.get_height()-4, "%"+str(round(p.get_height()/tips.day.count()*100, 2)));
```
